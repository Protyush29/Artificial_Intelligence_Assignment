```python
import numpy as np
from numpy import *
np.set_printoptions(threshold=np.inf)
from Queue import Queue
import time
```

```python
class Node(object):
    def __init__(self,parent=None,val=()):
        self.parent=parent
        self.val=val
    def get_val(self):
        return self.val
    def get_parent(self):
        return self.parent
```

```python
#Getting Inputs Ready
def get_content(path):
        with  open(path) as f:
            content = f.readlines()

        content= [list(x.strip()) for x in content]
        map_content = np.asanyarray(content)
        return map_content
```

```python
#finding the path
def get_path(goal,map_content,out):
        
        node=goal.get_parent()
        while(True):
            
            pos=node.get_val()
            if map_content[pos] != 's':
                map_content[pos]='X'
                node=node.get_parent()
                
            else: break
                
            np.savetxt(out+'path.txt', map_content,fmt= '%c',delimiter="")
            f = open(out+'.txt', 'r')
            file_contents = f.read()
            print (file_contents)
            f.close()
```
